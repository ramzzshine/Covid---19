```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['figure.figsize'] = (14, 5)
mpl.rcParams['axes.grid'] = False
```

```python
corona_confirmed = pd.read_csv('covid_confirmed_daily_updates.csv')
```

```python
corona_confirmed.head()
```

```python
corona_death_daily = pd.read_csv('covid_deaths_daily_updates.csv')
corona_death_daily.head()
```

```python
corona_recovered = pd.read_csv('covid_recovered_daily_updates.csv')
corona_recovered.head()
```

```python
corona_confirmed=corona_confirmed.melt(id_vars=['Country/Region'])
```

```python
corona_confirmed.head()
```

```python
corona_confirmed['value'].unique()
```

```python
corona_confirmed=corona_confirmed.rename({'variable':'Date','value': 'Confirmed'}, axis='columns')
```

```python
corona_confirmed.head()
```

```python
corona_death_daily=corona_death_daily.melt(id_vars=['Country/Region'])
```

```python
corona_death_daily['value'].unique()
```

```python
corona_death_daily=corona_death_daily.rename({'variable':'Date','value': 'Death'}, axis='columns')
```

```python
corona_death_daily.head()
```

```python
corona_recovered=corona_recovered.melt(id_vars=['Country/Region']).rename({'variable':'Date','value': 'Recovered'}, axis='columns')
```

```python
corona_recovered.head()
```

```python
combined_dataset = [corona_confirmed, corona_death_daily, corona_recovered]
combined_dataset = [df.set_index(['Country/Region','Date']) for df in combined_dataset]
combined_dataset=combined_dataset[0].join(combined_dataset[1:])
```

```python
combined_dataset.head(10)
```

```python
combined_dataset.dtypes
```

```python
combined_dataset=combined_dataset.reset_index()
```

```python
combined_dataset.head()
```

```python
#combined_dataset[['Confirmed','Death','Recovered']] = combined_dataset[['Confirmed','Death','Recovered']].apply(pd.to_numeric)
```

```python
combined_dataset[['Date']] = combined_dataset[['Date']].apply(pd.to_datetime)
```

```python
combined_dataset.dtypes
```

```python
combined_dataset=combined_dataset.rename({'Country/Region': 'Country'}, axis='columns')
```

```python
combined_dataset.columns
```

```python
combined_dataset.query("Country=='US'")
```

```python
combined_dataset['Country'].unique()
```

```python
combined_dataset.query("Country=='Mainland China'")
```

```python
combined_dataset['Active']=combined_dataset['Confirmed'] - combined_dataset['Death'] - combined_dataset['Recovered']
```

```python
combined_dataset.query("Country=='Mainland China'")
```

```python
Max_Cases=combined_dataset[combined_dataset.Date==combined_dataset['Date'].max()]
```

**Country by Country Max cases**

```python
Max_Cases[['Country','Active']].sort_values('Active', ascending = False).head(10)
```

```python
Max_Cases[['Country','Confirmed']].groupby(['Country']).sum().sort_values('Confirmed',ascending=False).head(10)
```
